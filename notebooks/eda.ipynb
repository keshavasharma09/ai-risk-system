```python
df = load_data("transactions.csv")
df.head()
```

```python
df["Class"].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df["Amount"] = sc.fit_transform(df[["Amount"]])
df["Amount"]
```

```python
df["Class"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, :-1]
y= df["Class"]
print(df.shape, X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=2)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Initialize SMOTE
# SMOTE, by default, equals to the number of samples in all classes
smote = SMOTE(random_state=2)

# Fitting and transforming training data
X_smote, y_smote = smote.fit_resample(X_train, y_train)

print(f"Original distribution: {Counter(y_train)}")
print(f"SMOTE distribution: {Counter(y_smote)}")
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_smote, y_smote)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report, confusion_matrix

y_pred = lr.predict(X_test)

print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print("Recall Score: ", recall_score(y_test, y_pred))
print("Precision Score: ", precision_score(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
print("Classification Report: ", classification_report(y_test, y_pred))
print("COnfusion Matrix:\n ", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(criterion="entropy")
rf.fit(X_smote, y_smote)

y_pred = rf.predict(X_test)

print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print("Recall Score: ", recall_score(y_test, y_pred))
print("Precision Score: ", precision_score(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
print("Classification Report: ", classification_report(y_test, y_pred))
print("COnfusion Matrix:\n ", confusion_matrix(y_test, y_pred))
```

```python
from imblearn.combine import SMOTEENN
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve, f1_score
import numpy as np

# 1. Apply SMOTE-ENN to Training Data ONLY
sme = SMOTEENN(random_state=42)
X_res, y_res = sme.fit_resample(X_train, y_train)

# 2. Train Model
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_res, y_res)

# 3. Get Probabilities for the Minority Class (Class 1)
# Use the Test set here to see how it performs on real-world data
y_probs = rf.predict_proba(X_test)[:, 1]

# 4. Calculate Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

# 5. Find the Optimal Threshold (Maximizing F1-Score)
# We calculate F1 for every threshold provided by the curve
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
ix = np.argmax(f1_scores)
best_threshold = thresholds[ix]

print(f'Best Threshold: {best_threshold:.4f}')
print(f'Best F1-Score: {f1_scores[ix]:.4f}')

# 6. Apply the New Threshold
y_pred_optimal = (y_probs >= best_threshold).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_optimal))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import precision_recall_curve
import numpy as np

# Calculating ration for ideal scale_pos_weight
ratio = int(np.sqrt(df["Class"].value_counts()[0] / df["Class"].value_counts()[1]))

xgb = XGBClassifier(
    scale_pos_weight=ratio, 
    max_delta_step=1,
    learning_rate=0.1,
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
# No need for SMOTE-ENN with scale_pos_weight
xgb.fit(X_train, y_train)
xgb_y_prob = xgb.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, xgb_y_prob)
f1_score = 2*(precision * recall) / (precision + recall + 1e-10)
best_index = np.argmax(f1_score)
best_threshold = thresholds[best_index]

print(f"Optimal Threshold: {best_threshold:.4f}")
print(f"Best F1-Score: {f1_score[best_index]:.4f}")

y_pred_optimal = (y_probs >= best_threshold).astype(int)
```
